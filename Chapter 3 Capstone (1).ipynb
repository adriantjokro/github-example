```python
r = requests.get('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M')
```

```python
soup = BeautifulSoup(r.text, 'lxml')
```

```python
print(soup.title)
```

```python
tab = str(soup.table)
```

```python
tab
```

```python
df1 = pd.read_html(tab)
df2=df1[0]
df2.head()
```

```python
dfdf3 = df2.where(df2.Borough != 'Not assigned')
```

```python
df4 = df3.dropna()
```

```python
df4.shape
```

```python
df5 = df4.groupby(['Postal Code'], sort=False).agg(', '.join)
df5.reset_index(inplace=True)
dfloc = pd.read_csv('https://cocl.us/Geospatial_data')
dfloc.head()
```

```python
df6 = pd.merge(df5,dfloc,on='Postal Code')
df6.head()
```

```python



# create map of New York using latitude and longitude values
map_toronto = folium.Map(location=[43.653, -79.3606], zoom_start=10)

# add markers to map
for lat, lng, borough, neighborhood in zip(df6['Latitude'], df6['Longitude'], df6['Borough'], df6['Neighborhood']):
    label = '{}, {}'.format(neighborhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_toronto)  
    
map_toronto
```
